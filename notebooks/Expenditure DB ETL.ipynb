```python
#importing needed libs
import pandas as pd
```

```python
#reading the raw .HTML DB
DATA = pd.read_html('../raw data/expenditure.html')
DATA
```

```python
#turning the 'DATA' list object into a proper Pandas DF
df = pd.concat(DATA)
df
```

```python
df.head(5)
```

```python
#dropping 'VAT' row since it's only a type of Tax and it isn't really relevant to the project
df.drop(4, axis=0, inplace=True)
df
```

```python
#cleaning the unnecessary and wrong characters from the numeric data columns
cols_to_clean = ['This Month',
                'Last Month',
                'This Season',
                'Last Season']
#making a for loop to clean all columns at once
for col in df.columns:
    df[col] = df[col].str.replace('Â£','')
df.head(5)
```

```python
#removing the comma from numeric data in the dataframe
for col in df.columns:
    df[col] = df[col].str.replace(',','')
df.head(5)
```

```python
#adding a date column to the dataframe
date = '6/28/2021'

df['Date'] = date
```

```python
#renaming the 'Item' column for better understanding
df.rename(columns={'Item':'Source'}, inplace=True)
df
```

```python
df.to_csv('../treated data/expenditure.csv', index=False)
```
