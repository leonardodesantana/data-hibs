```python
import pandas as pd
```

```python
#getting the data to be treated
pd.options.display.max_rows = 999
dfs = pd.read_csv('C:/data-hibs/data/FIFA_countries_db.csv')
dfs.head(5)
```

```python
#getting rid of the first row of the dataframe
dfs.drop(0)

#setting a new dataframe to work proper
df = pd.DataFrame({'Acronym' : dfs['0'], 
                   'Name' : dfs['1'],
                   'Acronym1': dfs['2'], 
                   'Name1': dfs['3'],
                  'Acronym2': dfs['4'],
                  'Name2': dfs['5']})

#wiping the first row (old column titles)
df.drop(0)
df.head(5)
```

```python
#creating a acronyms only dataframe
acronym = pd.DataFrame(df['Acronym'])
acronym1 = pd.DataFrame(df['Acronym1'])
acronym2 = pd.DataFrame(df['Acronym2'])

#cleaning unnecessary rows, rows_to_drop2
rows_to_drop = [0, 72]
rows_to_drop2 = [0,70,71,72]
acronym = acronym.drop(rows_to_drop)
acronym1 = acronym1.drop(rows_to_drop)
acronym2 = acronym2.drop(rows_to_drop2)

#setting all dataframes with the same column name, making the pd.concat work properly
acronym.columns = ['Acronym']
acronym1.columns = ['Acronym']
acronym2. columns = ['Acronym']
```

```python
#all acronyms are inserted in the same dataframe
combined_df = pd.concat([acronym, acronym1, acronym2], axis=0, ignore_index=True)
acronymdf = combined_df
```

```python
#creating countries name dataframes for further concat

df_name = pd.DataFrame(dfs['1'])
df_name2 = pd.DataFrame(dfs['3'])
df_name3 = pd.DataFrame(dfs['5'])

#unnecessary rows cleaned
rows_to_drop = [0,72]
rows_to_drop2 = [0,70,71,72]
df_name = df_name.drop(rows_to_drop)
df_name2 = df_name2.drop(rows_to_drop)
df_name3 = df_name3.drop(rows_to_drop2)

#all the columns put on the same name to map the pd.concat properly
df_name.columns = ['Name']
df_name2.columns = ['Name']
df_name3. columns = ['Name']

#all countries names are now put in the same dataframe
data = pd.concat([df_name, df_name2, df_name3], axis=0, ignore_index=True)
```

```python
#countries' names and acronyms properly setted into the same dataframe

merged_df = acronymdf.merge(data, left_index=True, right_index=True)
merged_df
```

```python
merged_df['Country ID'] = merged_df.reset_index().index+1
merged_df.to_csv('C:/data-hibs/data/countries.csv', index=False)
```
