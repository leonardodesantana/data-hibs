```python
#importing necessary libraries
import pandas as pd
import html
```

```python
#reading the .html db exported from Football Manager 2022 print feature
html_db = pd.read_html('C:\data-hibs\data\players.html')
html_db
```

```python
#turning the list into a proper Pandas Data Frame
db_to_pdDF = pd.concat(html_db)
df = pd.DataFrame(db_to_pdDF)
df
```

```python
#treating some wrong data in the 'Name' column
player_UID = 96075402
df.loc[df['UID'] == player_UID, 'Name'] = 'Kevin Dąbrowski'

player_UID2 = 36062516
df.loc[df['UID'] == player_UID2, 'Name'] = 'Alex Gogić'
df.head(5)
```

```python
df.to_csv('../treated data/players.csv')
```
