```python
#importing needed libraries

import pandas as pd
```

```python
#getting the players.csv

df = pd.read_csv('../treated data/players.csv')
df.head(5)
```

```python
#removing the 'cm' and 'kg' sufix from the columns Height and Weight
df['Height'] = df['Height'].str.replace(" cm", "")
df['Weight'] = df['Weight'].str.replace(' kg', '')
df.head(6)
```

```python
#getting rid of this column that popped from god knows where
del df['Unnamed: 0']
```

```python
#saving the changes into the .csv file again
df.to_csv('../treated data/players.csv')
```
