```python
#getting necessary libraries

import pandas as pd
```

```python
#reading the .csv into a pandas dataframe
df = pd.read_csv('../treated data/players.csv')
df.head(4)
```

```python
#getting rid of index column
columns_to_drop = ['Unnamed: 0']
df = df.drop(columns=columns_to_drop)
df.head(5)
```

```python
#assign the country.csv as a pandas dataframe
country_df = pd.read_csv('../treated data/countries.csv')

#changing the 'Acronym' column name to 'Nat', so it can be merged further
country_df = country_df.rename({'Acronym': 'Nat'}, axis='columns')
country_df.head(5)
```

```python
#making a left join of the Country ID to the players.csv table
leftjoin_df = pd.merge(df, 
                    country_df,
                    on='Nat',
                    how='left')
leftjoin_df.head(5)
```

```python
#getting rid of country 'dimensional' data
columns_to_drop = ['Nat','Name_y']
df = leftjoin_df.drop(columns=columns_to_drop)
```

```python
#making the same process for the '2nd Nat' (2nd nationality) in the df

#changing the columns name to make the pd.merge work propertly
df = df.rename({'Name_x': 'Player Name', '2nd Nat': 'Nat'}, axis='columns')

#doing the merged dataframe
merged_df = pd.merge(df,
                    country_df,
                    on='Nat',
                    how='left')
```

```python
#getting rid of dimensional columns again
columns_to_drop = ['Name', 'Nat']
merged_df = merged_df.drop(columns=columns_to_drop)
```

```python
#Changing columns names to make things clear
merged_df = merged_df.rename({'Country ID_x': 'Country ID',
                              'Country ID_y': '2nd Nat Country ID'}, 
                             axis='columns')
```
