```python
#getting the necessary libs
import pandas as pd
```

```python
#turning the raw .HTML DB into a python object
data = pd.read_html('../raw data/training.html')

#now turning the object in a proper pandas dataframe
df = pd.concat(data)
df
```

```python
#getting a list of the dataframe structure
for col in df.columns:
    print("{}: {}".format(col,df[col].dtype))
```

```python
#getting rid of the 'Name' column
column_to_drop = ['Name']

df.drop(columns=column_to_drop, inplace=True)
```

```python
for col in df.columns:
    print("{}: {}".format(col,df[col].dtype))
```

```python
#changing the 'GKS' column data type from bool to object
df['GKS'] = df['GKS'].astype('object')

print(df.dtypes)
```

```python
#dropping unnecessary columns
columns_to_drop = ['AtTa','DeTe','Fit','GKH','GKS','PoTa','PoTe']

df.drop(columns=columns_to_drop, inplace=True)

df.head(5)
```

```python
training_date = '6/28/2021'

df['Week of the Session'] = training_date
df
```

```python
df.to_csv('../treated data/training.csv', index=False)
```
